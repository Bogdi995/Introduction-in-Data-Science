6. (2p) A dictionary is given in which the keys are names of students and the values are lists of favourite books. Determine:
    1. Which students have a specified book in their list of favourites?
    2. Which pairs of students have the same book preferences (the order in the preference list is not relevant).
    3. For a given student, which are the students with whom he/she has the most common books in his/her preference list; if there are several such students closest together, list them all.
    4. For two students, A and B, what is the list of deletions and additions by which A's list becomes identical to B's list (the order in the list is not important).
    5. All pairs of students A, B for which A's preference list is included in (but does not coincide with) B's preference list. 
    
Hint: consider the data type set.

Example:
```python
preferences = {
    'Popescu': ['book2', 'book1', 'book3', 'book4'],
    'Ionescu': ['book1', 'book2', 'book3', 'book7', 'another book'],
    'Georgescu': ['book1', 'another book'],
    'Xulescu': ['book2', 'book5', 'book6', 'book7'],
    'Dragomir': ['book4', 'book1', 'book2', 'book3']
}
```
    1. book = 'another book' -> ['Ionescu', 'Georgescu']
    2. Popescu and Dragomir have the same preferences: ['book2', 'book1', 'book3', 'book4']. Note that the equivalent pair is not displayed: Dragomir and Popescu.
    3. Xulescu -> Maximum overlap of preferences: 2 with peers: ['Ionescu']
    4. A='Popescu', B='Ionescu' -> Add books: {'book7', 'another book'}, remove books: {'book4'}
    5. Georgescu's preferences are a subset of Ionescu's preferences.
    


````python
import collections

preferences = {
    'Popescu': ['book2', 'book1', 'book3', 'book4'],
    'Ionescu': ['book1', 'book2', 'book3', 'book7', 'another book'],
    'Georgescu': ['book1', 'another book'],
    'Xulescu': ['book2', 'book5', 'book6', 'book7'],
    'Dragomir': ['book4', 'book1', 'book2', 'book3']
}

def students_with_specified_book(book, preferences):
    students = []
    for key in preferences:
        for item in preferences[key]:
            if item == book:
                students.append(key)
    print(students)

    
def students_with_same_preferred_books(preferences):
    students = []
    for firstKey in preferences:
        for secondKey in preferences:
            if(firstKey != secondKey):
                if(collections.Counter(preferences[firstKey]) == collections.Counter(preferences[secondKey])):
                    if(firstKey not in students):
                        students.append(firstKey)
                    if(secondKey not in students):
                        students.append(secondKey)
    print(students)

    
def get_books_from_student(name):
    student = []
    for key in preferences:
        if key == name:
            return preferences[key]
    
    
def students_with_most_common_books(name):
    students = []
    maxNumber = 0
    books = get_books_from_student(name)
    for key in preferences:
        if key != name:
            if(maxNumber < len(set(books) & set(preferences[key]))):
                students.clear();
                students.append(key)
                maxNumber = len(set(books) & set(preferences[key]))
            elif(maxNumber == len(set(books) & set(preferences[key]))):
                students.append(key)
    print(students)
    
    
def equalize_the_preferred_books(student1, student2):
    differences = []
    student1Books = get_books_from_student(student1)
    student2Books = get_books_from_student(student2)
    willBeAdded = []
    willBeEliminated = []
    differences = list(list(set(student1Books) - set(student2Books)) + list(set(student2Books) - set(student1Books)))
    
    for item in differences:
        if item not in student1Books:
            willBeAdded.append(item)
            
    for item in differences:
        if item not in student2Books:
            willBeEliminated.append(item)
            
    print('The following books will be added: ' +  str(willBeAdded))
    print('The following books will be eliminated: ' + str(willBeEliminated))
    
    
def list_of_books_included_in_another_list(name, preferences):
    studentBooks = get_books_from_student(name)
    for key in preferences:
        if key != name:
            if set(studentBooks).issubset(preferences[key]):
                print(f"{name} preferences are a subset of {key}'s preferences.")
````

````python
givenBook = 'another book'
givenStudent1 = 'Xulescu'
givenStudent2 = 'Popescu'
givenStudent3 = 'Ionescu'
givenStudent4 = 'Georgescu'
students_with_specified_book(givenBook, preferences)
students_with_same_preferred_books(preferences)
students_with_most_common_books(givenStudent1)
equalize_the_preferred_books(givenStudent2, givenStudent3)
list_of_books_included_in_another_list(givenStudent4, preferences)
````
