```python
[
 {"name": "Popescu",
  "gender": "feminine",
  "age": 30,
  "salary": 2000,
  "favorite animal": "cat",
 },
 {"name": "Irimia",
  "age": 40,
  "gender": "male"
  "salary": 3000,
  "pet": "dog"
 },
    # etc
]
```
To determine:<br>
&nbsp;&nbsp;&nbsp;&nbsp;1. Average, minimum, and maximum wage<br>
&nbsp;&nbsp;&nbsp;&nbsp;2. Those who are of a given age in a given interval [a, b]<br>
&nbsp;&nbsp;&nbsp;&nbsp;3. Which animal has the most preferences. If there are several such animals, then all of them will be given<br>
&nbsp;&nbsp;&nbsp;&nbsp;4. How many males and females are in the l
    
Notes: All dictionaries have the same keys: names etc. 

````python
import statistics
from collections import Counter

list = [
    {
        "name": "Popescu",
        "gender": "female",
        "age": 30,
        "salary": 2000,
        "favorite animal": "cat",
    },
    
    {
        "name": "Irimia",
        "age": 40,
        "gender": "male",
        "salary": 3000,
        "favorite animal": "dog"
    },

    {
        "name": "Ionescu",
        "gender" : "male",
        "age": 25,
        "salary": 1300,
        "favorite animal": "dog"
    },
    
    {
        "name": "Blaga",
        "gender" : "male",
        "age": 20,
        "salary": 2500,
        "favorite animal": "dog"
    },
    
    
    {
        "name": "Rusu",
        "gender" : "female",
        "age": 20,
        "salary": 2800,
        "favorite animal": "cat"
    },
    
    {
        "name": "Negrea",
        "gender" : "female",
        "age": 32,
        "salary": 6300,
        "favorite animal": "hamster"
    },
    
    {
        "name": "Stegaru",
        "gender" : "male",
        "age": 68,
        "salary": 3400,
        "favorite animal": "hamster"
    }
]
````

````python
def average_salary(list):
    allSalaries = [dict['salary'] for dict in list]
    return statistics.mean(allSalaries)

def maximum_salary(list):
    allSalaries = [dict['salary'] for dict in list]
    return max(allSalaries)

def minimum_salary(list):
    allSalaries = [dict['salary'] for dict in list]
    return min(allSalaries)

def age_in_interval(list, a, b):
    return [dict['name'] for dict in list if dict['age'] in range(a,b)]

def favorite_animal(list):
    animalList = [dict["favorite animal"] for dict in list]
    c = Counter(animalList)
    animalListSet = set(animalList)
    maxCount = -1
    for animal in animalListSet:
        if(c[animal] > maxCount):
            maxCount = c[animal]
            
    return [animal for animal in animalListSet if c[animal] == maxCount]

def males_number(list):
    return len([dict["name"] for dict in list if dict["gender"] == "male"])

def females_number(list):
    return len([dict["name"] for dict in list if dict["gender"] == "female"])
````

````python
a = 20
b = 30

print("The average salary is:", average_salary(list))
print("The maximum salary is:", maximum_salary(list))
print("The minimum salary is:", minimum_salary(list))
print("The persons between", a, b, "are", age_in_interval(list, a, b))
print(favorite_animal(list))
print("The number of males is:", males_number(list))
print("The number of females is:", females_number(list))
````
