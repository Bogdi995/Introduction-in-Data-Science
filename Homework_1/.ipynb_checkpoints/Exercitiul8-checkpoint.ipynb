```python
import statistics
from collections import Counter

lista = [
    {
        "nume": "Popescu",
        "gen": "feminin",
        "varsta": 30,
        "salariu": 2000,
        "animal preferat": "pisica",
    },
    
    {
        "nume": "Irimia",
        "varsta": 40,
        "gen": "masculin",
        "salariu": 3000,
        "animal preferat": "caine"
    },

    {
        "nume": "Ionescu",
        "gen" : "masculin",
        "varsta": 25,
        "salariu": 1300,
        "animal preferat": "caine"
    },
    
    {
        "nume": "Blaga",
        "gen" : "masculin",
        "varsta": 20,
        "salariu": 2500,
        "animal preferat": "caine"
    },
    
    
    {
        "nume": "Matrache",
        "gen" : "masculin",
        "varsta": 20,
        "salariu": 2800,
        "animal preferat": "pisica"
    },
    
    {
        "nume": "Negrea",
        "gen" : "feminin",
        "varsta": 32,
        "salariu": 6300,
        "animal preferat": "hamster"
    },
    
    {
        "nume": "Stegaru",
        "gen" : "masculin",
        "varsta": 68,
        "salariu": 3400,
        "animal preferat": "hamster"
    }
]

def salariu_mediu(list):
    allSalaries = [dict['salariu'] for dict in list]
    return statistics.mean(allSalaries)

def salariu_maxim(list):
    allSalaries = [dict['salariu'] for dict in list]
    return max(allSalaries)

def salariu_minim(list):
    allSalaries = [dict['salariu'] for dict in list]
    return min(allSalaries)

def varsta_in_interval(list, a, b):
    return [dict['nume'] for dict in list if dict['varsta'] in range(a,b)]

def animalul_preferat(list):
    animalList = [dict["animal preferat"] for dict in list]
    c = Counter(animalList)
    animalListSet = set(animalList)
    maxCount = -1
    for animal in animalListSet:
        if(c[animal] > maxCount):
            maxCount = c[animal]
    return [animal for animal in animalListSet if c[animal] == maxCount]

def numar_sex_masculin(list):
    return len([dict["nume"] for dict in list if dict["gen"] == "masculin"])

def numar_sex_feminin(list):
    return len([dict["nume"] for dict in list if dict["gen"] == "feminin"])

a = 20
b = 30

print("Salariul mediu este:", salariu_mediu(lista))
print("Salariul maxim este:", salariu_maxim(lista))
print("Salariul minim este:", salariu_minim(lista))
print("Persoanele intre", a, b, "sunt", varsta_in_interval(lista,a,b))
print(animalul_preferat(lista))
print("Numarul de persoane de sex masculin este:", numar_sex_masculin(lista))
print("Numarul de persoane de sex feminin este:", numar_sex_feminin(lista))


```
