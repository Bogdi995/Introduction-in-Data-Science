```python
def IsVowel(character):
    if character in "aeiou":
        return True
    return False

def CheckBirdify(vowel,birdify):
    newWord1 = vowel + 'p' + vowel
    newWord2 = vowel + 'P' + vowel
    if newWord1 == birdify or newWord2 == birdify:
        return True
    return False

word = "bupucupu"

isBirdified = True
index = 0
while index in range(0, len(word) - 2):
    if IsVowel(word[index]) is True:
        birdify = word[index] + word[index + 1] + word[index + 2]
        if CheckBirdify(word[index], birdify) is False:
            isBirdified = False
            break
        index += 2
    index += 1
    
if(IsVowel(word[len(word)-2]) is True or (IsVowel(word[len(word)-3]) is False and IsVowel(word[len(word)-1]) is True)):
    isBirdified = False;

if isBirdified is True:
    print('Stringul provine din pasareasca')
else:
    print('Stringul nu provine din pasareasca')
```

```python
import collections
preferences = {
    'Popescu': ['carte2', 'carte1', 'carte3', 'carte4'],
    'Ionescu': ['carte1', 'carte2', 'carte3', 'carte7', 'alta carte'],
    'Georgescu': ['carte1', 'alta carte'],
    'Xulescu': ['carte2', 'carte5', 'carte6', 'carte7'],
    'Dragomir': ['carte4', 'carte1', 'carte2', 'carte3']
}

def StudentsWithSpecifiedBook(book,preferences):
    students = []
    for key in preferences:
        for item in preferences[key]:
            if item == book:
                students.append(key)
    print(students)

    
def StudentsWithSamePreferedBooks(preferences):
    students = []
    for firstKey in preferences:
        for secondKey in preferences:
            if(firstKey != secondKey):
                if(collections.Counter(preferences[firstKey]) == collections.Counter(preferences[secondKey])):
                    if(firstKey not in students):
                        students.append(firstKey)
                    if(secondKey not in students):
                        students.append(secondKey)
    print(students)

    
def GetBooksFromStudent(name):
    student = []
    for key in preferences:
        if key == name:
            return preferences[key]
    
    
def StudentsWithMostCommonBooks(name):
    students = []
    maxNumber = 0
    books = GetBooksFromStudent(name)
    for key in preferences:
        if key != name:
            if(maxNumber < len(set(books) & set(preferences[key]))):
                students.clear();
                students.append(key)
                maxNumber = len(set(books) & set(preferences[key]))
            elif(maxNumber == len(set(books) & set(preferences[key]))):
                students.append(key)
    print(students)
    
    
def EqualizeThePreferedBooks(student1, student2):
    differences = []
    student1Books = GetBooksFromStudent(student1)
    student2Books = GetBooksFromStudent(student2)
    willBeAdded = []
    willBeEliminated = []
    differences = list(list(set(student1Books)-set(student2Books)) + list(set(student2Books)-set(student1Books)))
    for item in differences:
        if item not in student1Books:
            willBeAdded.append(item)
    for item in differences:
        if item not in student2Books:
            willBeEliminated.append(item)
    print('The following books will be added: ' +  str(willBeAdded))
    print('The following books will be eliminated: ' + str(willBeEliminated))
    
    
def ListOfBooksIncludedInAnotherList(name, preferences):
    studentBooks = GetBooksFromStudent(name)
    for key in preferences:
        if key != name:
            if set(studentBooks).issubset(preferences[key]):
                print(f"{name} preferences are included in {key}'s preferences")


givenBook = 'alta carte'
givenStudent1 = 'Xulescu'
givenStudent2 = 'Popescu'
givenStudent3 = 'Ionescu'
givenStudent4 = 'Georgescu'
StudentsWithSpecifiedBook(givenBook, preferences)
StudentsWithSamePreferedBooks(preferences)
StudentsWithMostCommonBooks(givenStudent1)
EqualizeThePreferedBooks(givenStudent2, givenStudent3)
ListOfBooksIncludedInAnotherList(givenStudent4, preferences)
```

```python
def get_all_items(dictionary):
    myList = []
    for key, arrayValue in dictionary.items():
        for value in arrayValue:
            myList.append(value)
    return myList

def get_maximum_interval_apperances(c, all_items):
    maximum = -1
    for interval in set(all_items):
        if c[interval] > maximum:
            maximum = c[interval]
    return maximum
        
def print_result(maximum_intervals, doodle, keys):
    myDict = {}
    for interval in maximum_intervals:
        myDict[interval] = []
        for key in keys:
            if interval in doodle[key]:
                myDict[interval].append(key)
    groupList = []
    for key in myDict.keys():
        commonList = [k for k in myDict.keys() if myDict[key] == myDict[k]]
        if commonList not in groupList:
            groupList.append(commonList)
            names = [names for names in myDict[key]]
            print(names, "sunt disponibili in intervalele/intervalul", commonList)
        
                
    

from collections import Counter
doodle = {
"Popescu": ["13", "13.30", "16", "17"], 
"Ionescu": ["13", "13.30", "16", "17", "17.30", "11", "15"],
"Maximilian": ["14", "14.30", "15", "15.30", "16", "16.30", "19", "19.30"]
}


all_items = get_all_items(doodle)
c = Counter(all_items)
max_apperances = get_maximum_interval_apperances(c, all_items)
maximum_intervals = [key for key in set(all_items) if c[key] == max_apperances]

print_result(maximum_intervals, doodle, list([*doodle]))
```

```python
import statistics
from collections import Counter

lista = [
    {
        "nume": "Popescu",
        "gen": "feminin",
        "varsta": 30,
        "salariu": 2000,
        "animal preferat": "pisica",
    },
    
    {
        "nume": "Irimia",
        "varsta": 40,
        "gen": "masculin",
        "salariu": 3000,
        "animal preferat": "caine"
    },

    {
        "nume": "Ionescu",
        "gen" : "masculin",
        "varsta": 25,
        "salariu": 1300,
        "animal preferat": "caine"
    },
    
    {
        "nume": "Blaga",
        "gen" : "masculin",
        "varsta": 20,
        "salariu": 2500,
        "animal preferat": "caine"
    },
    
    
    {
        "nume": "Matrache",
        "gen" : "masculin",
        "varsta": 20,
        "salariu": 2800,
        "animal preferat": "pisica"
    },
    
    {
        "nume": "Negrea",
        "gen" : "feminin",
        "varsta": 32,
        "salariu": 6300,
        "animal preferat": "hamster"
    },
    
    {
        "nume": "Stegaru",
        "gen" : "masculin",
        "varsta": 68,
        "salariu": 3400,
        "animal preferat": "hamster"
    }
]

def salariu_mediu(list):
    allSalaries = [dict['salariu'] for dict in list]
    return statistics.mean(allSalaries)

def salariu_maxim(list):
    allSalaries = [dict['salariu'] for dict in list]
    return max(allSalaries)

def salariu_minim(list):
    allSalaries = [dict['salariu'] for dict in list]
    return min(allSalaries)

def varsta_in_interval(list, a, b):
    return [dict['nume'] for dict in list if dict['varsta'] in range(a,b)]

def animalul_preferat(list):
    animalList = [dict["animal preferat"] for dict in list]
    c = Counter(animalList)
    animalListSet = set(animalList)
    maxCount = -1
    for animal in animalListSet:
        if(c[animal] > maxCount):
            maxCount = c[animal]
    return [animal for animal in animalListSet if c[animal] == maxCount]

def numar_sex_masculin(list):
    return len([dict["nume"] for dict in list if dict["gen"] == "masculin"])

def numar_sex_feminin(list):
    return len([dict["nume"] for dict in list if dict["gen"] == "feminin"])

a = 20
b = 30

print("Salariul mediu este:", salariu_mediu(lista))
print("Salariul maxim este:", salariu_maxim(lista))
print("Salariul minim este:", salariu_minim(lista))
print("Persoanele intre", a, b, "sunt", varsta_in_interval(lista,a,b))
print(animalul_preferat(lista))
print("Numarul de persoane de sex masculin este:", numar_sex_masculin(lista))
print("Numarul de persoane de sex feminin este:", numar_sex_feminin(lista))
```
